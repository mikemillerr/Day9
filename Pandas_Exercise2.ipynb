Let's look at duration....

How long is the shortest trip? How many are that short?

```python
print(df['duration'].min())
print(df['duration'].max())
```

What do you think is going on with the shrort trips?

```python
print(df[df['duration'] == 60].shape[0])
short_trips = df[(df['start_station'] == df['end_station']) & (df['duration'] <= 60)]
```

There have been 17 trips with a duration below or equal to 60 seconds. Since all those trips end in the same station I'm assunming that those trips have been errors with the bike renting system. Those bikes have not traveled a distance in those cases.

What is the longest trip? How many trips are "long"?

```python
longest_trip = df['duration'].max() / 60
mean_trip = df['duration'].mean() / 60
median_trip = df['duration'].median() / 60
std_trip = df['duration'].std() / 60
print('Longest longest_trip in minutes %.2f' %longest_trip)
print('Mean trip in minutes %.2f' %mean_trip)
print('Median trip in minutes %.2f' %median_trip)
print('Std trip in minutes %.2f' %std_trip)
```

The longest trip took over 200 hours. I think in this particular trip an error has been made. Ether the person taking the bike forgot to give it back. Or there might have been an error in the trip finishing procedure.

```python
df.plot(y='duration', x='trip_id', kind='scatter');
```

We can see that the majority of trips have a maximal duration of around 100000 seconds. So to get a better view of the distribution, a histogram with a threshold of 120000 will give a better understanding of the duration.

```python
df_cut = df[df['duration'] <= 4000]
df_cut.shape[0]
num_cut = (df.shape[0] - df_cut.shape[0])
num_cut_per = (num_cut / df.shape[0]) * 100
print('Number of outliers cut from the dataframe: %.f2' % num_cut)
print('That amounts to %.2f prescent of the data' % num_cut_per )
```

```python
df_cut['duration'].plot(kind='hist', bins= 100);
```

```python
long_duration_treshold = (median_trip + 2* std_trip) * 60
df[df['duration'] >= long_duration_treshold].shape[0]
```

Setting the treshold for a long trip at median plus two standard deviations. There are 1681 longs trips!
