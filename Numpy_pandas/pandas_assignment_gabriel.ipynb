4) Let's say that I want to get started on cutting back next month. This time, though, I want to focus on staying away from those wines with a high acidity. Specifically, I want to stay away from those wines that are in the highest bin of fixed acidity (highest bin out of 5). You're now going to help me do this. To achieve this, let's do the following: 

 * Separate the rows in each `DataFrame` into 5 equal width bins based off their fixed acidity. 
 * Merge the resulting `Series` holding these 5 bins onto the original `DataFrame`. I can imagine also doing this with either `.join()` or using `pd.concat()`. Try doing it with both for practice. 
 * Return back to me all those rows that are **not in ** the top bin in terms of fixed acidity. 

```python
import numpy as np
ac_max = red_wine_df['fixed acidity'].max()
ac_min = red_wine_df['fixed acidity'].min()
```

```python
# Create a range object for the bins
#bin_width = (ac_max - ac_min) / 5
fixed_acidity_bins = np.linspace(ac_min, ac_max, 5)
fixed_acidity_bin_series_red = pd.cut(red_wine_df['fixed acidity'], bins=fixed_acidity_bins, labels=fixed_acidity_bins[:-1])
fixed_acidity_bin_series_red.name = 'fa_bin'
```

```python
# Join series and df with join
red_wine_df_ac_bined_red = red_wine_df.join(fixed_acidity_bin_series_red)

# Join series and df with concat
df_concated = pd.concat([red_wine_df, fixed_acidity_bin_series_red], axis=1)
```

```python
# Filter for wine which are not in the highest bin
red_wines_with_lower_ac = red_wine_df_ac_bined_red[red_wine_df_ac_bined_red['fa_bin']!= red_wine_df_ac_bined_red['fa_bin'].max()]
```
